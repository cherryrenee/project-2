```python
# 사용할 변수 선택
features = ['Price_USD', 'Size_ml', 'Number_of_Reviews', 'Skin_Type', 'Gender_Target', 'Usage_Frequency']
df_selected = df[features]

# 범주형 변수 원-핫 인코딩
df_encoded = pd.get_dummies(df_selected)

# 정규화 (KMeans는 거리 기반이므로 중요!)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_encoded)
```

```python
from sklearn.cluster import KMeans

# 군집 수 설정 (예: 4개 군집)
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# 군집별 평균 정보 확인
cluster_summary = df.groupby('Cluster')[['Price_USD', 'Size_ml', 'Number_of_Reviews']].mean()
cluster_counts = df['Cluster'].value_counts().sort_index()

cluster_summary['Count'] = cluster_counts
cluster_summary
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 군집별 평균값 계산 (수치형 변수만)
numeric_cols = ['Price_USD', 'Size_ml', 'Number_of_Reviews']
cluster_means = df.groupby('Cluster')[numeric_cols].mean()

# 히트맵 시각화
plt.figure(figsize=(8, 5))
sns.heatmap(cluster_means, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title("Cluster-wise Feature Averages (Heatmap)")
plt.ylabel("Cluster")
plt.xlabel("Feature")
plt.tight_layout()
plt.show()
```
